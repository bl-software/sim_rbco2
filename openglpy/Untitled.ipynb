```python
import matplotlib.pyplot as plt
from math import sin, cos, pi, sqrt, pow
import numpy as np
```

```python
D=7.7
r=np.linspace(0,D/2,20)
r
```

```python
rp2= np.power(r,2)
rp4= np.power(r,4)
dp2= np.power(D,2)
dp4= np.power(D,4)
plt.plot(r,rp2,r,rp4)
print(f'dp2:{dp2} dp4:{dp4}')
```

```python
rp2Odp2= rp2/dp2
plt.plot(r,  rp2Odp2,label='rp2 over dp2')
plt.plot(r,4*rp2Odp2,label='4 rp2 over dp2')
rp4Odp4= rp4/dp4
plt.plot(r,r*rp4Odp4,label='rp4 over dp4')
plt.legend()
```

```python
s=np.sqrt(1-4*rp2Odp2)
plt.plot(r,s,label='sqrt')
```

```python
from ipywidgets import interact, FloatSlider
def bicon(a0=0.02,a1=1.54,a2=2.2):
    a0part=a0 * np.ones(len(r))
    a1part=a1 * rp2Odp2
    a2part=a2 * rp4Odp4

    s=np.sqrt(1-4*rp2Odp2)

    fig, axs = plt.subplots(figsize=[16,20])
    axs=[axs,None]
    axs[0].plot(r,s,label='sqrt')
    axs[0].plot(r,a0part,label='a0part')
    axs[0].plot(r,a1part,label='a1part')
    axs[0].plot(r,a2part,label='a2part')
    #axs[0].legend()

    atot= a0part + a1part + a2part
    axs[0].plot(r,atot,label='atot')
    axs[0].plot(r,s,label='sqrt')
    bi=s*atot
    axs[0].plot(r,bi,label='bi')
    axs[0].legend()

interact(bicon,
        a0=FloatSlider(min=-3.2, max=3.2, step=0.2, value=0.0),
        a1=FloatSlider(min=-3.2, max=3.2, step=0.2, value=0.0),
        a2=FloatSlider(min=-3.2, max=3.2, step=0.2, value=0.0),
        )
```
